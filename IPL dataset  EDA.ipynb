```python

dfBall = Ball_by_ball.groupby('Striker')\
            .agg({'MatcH_id':'nunique','Runs_Scored':'sum','Ball_id':'count','Wides':'sum','Legbyes':'sum','Byes':'sum','Noballs':'sum'})\
            .reset_index()
```

```python
dfBall
```

```python
dfBall.columns = ['Batter','Innings','Total_Runs','total_ball','wides','legbyes','byes','noball']
```

```python
dfBall
```

## merging and adding new column in for batter 

```python
dfBall['ball_faced']=dfBall['total_ball']-dfBall['wides']-dfBall['legbyes']-dfBall['byes']-dfBall['noball']
```

```python
dfBall
```

```python
dfBall.drop(['wides','legbyes','byes','noball','total_ball'],axis=1)
```

```python
dfBall['batting_strike_rate']=((dfBall['Total_Runs']/dfBall['ball_faced'])*100).round(decimals=2)
```

```python
dfBall
```

```python
## total number of type player out
dfout = Ball_by_ball[(Ball_by_ball['Out_type']!="Not Applicable" )].groupby('Striker')['MatcH_id'].count().reset_index()
```

```python
dfout
```

```python
dfout.columns=['Batter','out_inning']
```

```python
dfout
```

```python
dfBall2=dfBall.merge(dfout,on='Batter',how='left')
```

```python
dfBall2
```

```python
dfBall2['batting_avg']=(dfBall2['Total_Runs']/dfBall2['out_inning']).round(decimals=2)
```

```python
dfBall2
```

```python
dfBall2.drop(['wides','legbyes','byes','noball','total_ball'],axis=1,inplace=True)
```

```python
dfBall2
```

```python
## dfball3 is merge of dfball2 and strike dataset
dfBall3=dfBall2.merge(df1[['Batter','player_name','player_country']],on='Batter',how='left')
```

```python
dfBall3
```

```python
dfBall4=dfBall3.merge(df3[['Batter','runs_from_fours','runs from sixes']],on='Batter',how='left')
```

```python
dfBall4
```

```python
dfBall4['boundry_run%']=(((dfBall4['runs_from_fours']+dfBall4["runs from sixes"])/dfBall4['Total_Runs'])*100).round(decimals=2)
```

```python
dfBall4
```

```python
dfBall4.to_csv('player_batting_dataset')
```

## We use the same aggregation methods with a combination of conditional slicing by excluding all the wide balls for calculating the number of balls faced, by including all the runs which are 4s and boundaries for calculating the number of 4s scored and similarly for 6s.

```python
dfdotball = Ball_by_ball[(Ball_by_ball['Runs_Scored']!= 0)].groupby('Bowler')['MatcH_id'].count().reset_index()
```

```python
dfdotball
```

```python
dfdotball.columns=['Bowler','Non_dotball']
```

```python
dfdotball
```

```python
dfBowling = Ball_by_ball.groupby('Bowler').agg({'MatcH_id':'nunique','Runs_Scored':'sum','Bowler_Wicket':'sum','Extra_runs':'sum','Ball_id':'count'}).reset_index()
```

```python
dfBowling
```

```python
dfBowling.columns =['Bowler','bowl_inning','Runs_conceded','bowler_wicket','bowler_extraruns','ball_throw']
```

```python
dfBowling
```

## merging dfdotball and dfBowling and adding new column

```python
dfBowling2=dfBowling.merge(dfdotball,on='Bowler',how='left')
```

```python
dfBowling2
```

```python
dfBowling2['dot_ball %']=(((dfBowling2['ball_throw']-dfBowling2['Non_dotball'])/dfBowling2['ball_throw'])*100).round(decimals=2)
```

```python
dfBowling2
```

```python
dfBowling2['bowling_avg']=(dfBowling2['Runs_conceded']/dfBowling2['bowler_wicket']).round(decimals=2)
```

```python
dfBowling2
```

```python
dfBowling2['bowling_economy']=(dfBowling2['Runs_conceded']/(dfBowling2['ball_throw']/6)).round(decimals=2)
```

```python
dfBowling2
```

```python
dfBowling2['bowling_strikerate']=(dfBowling2['ball_throw']/dfBowling2['bowler_wicket']).round(decimals=2)
```

```python
dfBowling2
```

```python
dfBowling2['Batter']=dfBowling['Bowler']
```

```python
dfBowling2.to_csv('player_bowling_dataset')
```
